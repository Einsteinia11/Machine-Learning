```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.arange(10)
print(x)
```

```python
plt.plot(x)# when supplied single array it plots arg on y and indices on x
plt.show()
```

```python
x=[2,4,6,8,10]
plt.plot(x)
plt.show()
```

```python
x=np.arange(4)
y=x**2
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
x=np.arange(4)
y=x**2
plt.plot(x, color="green")
plt.show()
plt.plot(y)
plt.show()
```

```python
plt.scatter(x,y,color="orange", label="stock price", marker="^") #color, label, marker are optional arguments
plt.legend()
plt.xlabel("days")
plt.ylabel("price")
plt.title("stock price prediction")
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([10,20,50,80,90])
plt.plot(x,x**2, label="2018")
plt.plot(x,x**3, label="2017")
plt.legend()
plt.xlabel("no of days")
plt.ylabel("stock price")
plt.title("stock prices")
plt.show()
```

### barcharts

```python
numbers=np.random.randint(0,10,5)
indices=np.arange(5)
numbers2=np.random.randint(0,10,5)
indices2=np.arange(5)
print(numbers)
print(numbers2)
```

```python
plt.bar(indices,numbers,0.25,color="red",label="gold price")
plt.bar(indices2+0.25,numbers2,0.25,color="green",label="silver price")
plt.legend()
plt.show()
```

### piechart

```python
labels=["algebra","calculus","coding","algorithm"]
values=[90,80,40,73]
plt.pie(values,labels=labels,radius=1)
plt.show()
plt.pie(values,labels=labels,radius=1,explode=(0,0,1,0),shadow=True)
plt.show()
```

### Histogram

```python
u=5
sigma=2
vals=u+sigma*np.random.randn(1000)
print(vals.shape)
plt.hist(vals,50)
plt.show()
```
